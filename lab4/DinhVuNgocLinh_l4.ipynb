````python

import sys

current_item = None
current_total_cost = 0.0

for line in sys.stdin:
    line = line.strip()
    item, cost = line.split("\t")

    try:
        cost = float(cost)
    except ValueError:
        continue
    if current_item == None:
        current_item = item
        current_total_cost = cost
    elif current_item == item:
        current_total_cost += cost
    else:
        print(f"{current_item}\t{current_total_cost}")
        current_item = item
        current_total_cost = cost

if current_item is not None:
    print(f"{current_item}\t{current_total_cost}")
````

````python

import sys

current_location = None
max_item = None
max_cost = float('-inf')

for line in sys.stdin:
    line = line.strip()
    location, item, cost = line.split("\t")
    cost = float(cost)
    if current_location is None or current_location != location:
        if current_location is not None:
            print(f"{current_location}\t{max_item}\t{max_cost}")
        current_location = location
        max_item = item
        max_cost = cost
    else:
        if cost > max_cost:
            max_item = item
            max_cost = cost

if current_location is not None:
    print(f"{current_location}\t{max_item}\t{max_cost}")
````
