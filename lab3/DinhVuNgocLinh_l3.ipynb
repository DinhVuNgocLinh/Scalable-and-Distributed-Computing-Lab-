## Exercise 1
Write a function to compute the below formualation with the use of map function
\begin{equation}
x^2 + y^2 - r^2
\end{equation}

then, print out the values of the equation at these below values of $x$, $y$ and $z$
```
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
r = [1,1,3,3,4,5]

```

````python
def compute_equation(x, y, r):
  results = list(map(lambda a, b, c: a**2 + b**2 - c**2, x, y, r))
  return results
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
r = [1,1,3,3,4,5]
print(compute_equation(x,y,r))
````

## Exercise 2

Write a function that take input as a list of words and return to a list of integers representing the lengths of the correponding words.

Write it in three different ways:
1. using a for-loop,
2. using function `map()`, and
3. using list comprehensions.

````python
def use_for_loop (words):
  length = []
  for i in words:
    length.append(len(i))
  return length

def use_map (words):
  return list(map(len,words))

def use_list_comprehensions(words):
  return [len(i) for i in words]
````

## Exercise 3

Implement a program to make a list whose elements are double string of a string `re` (i.e., `rere`), where `re` is the first two characters of approriate elements in this list:

`['real', 'rock', 'realize', 'rocket', 'red', 'eraser' , '7', 8 , 9, 10]`.

````python
words = ['real', 'rock', 'realize', 'rocket', 'red', 'eraser', '7', 8, 9, 10]

double_strings = [i[:2]*2 for i in words if isinstance(i, str)]

print(double_strings)
````

## Exercise 4

Write a function `max_in_list()` that takes a list of numbers and returns the largest one using `reduce()` in lib `functools`

````python
from functools import reduce

def max_in_list(list):
  return reduce(lambda a, b: a if a > b else b, list)
````

## Exercise 5

Using Python `map` to write a Python program to triple all numbers of a given list of integers.

````python
def triple_numbers(nums):
    tripled_nums = list(map(lambda x: x * 3, nums))
    return tripled_nums
````

## Exercise 6
Implement parallel processing using `map()` to apply a function across multiple elements of an iterable concurrently, and then use `reduce()` to aggregate the results.

````python
import multiprocessing
from functools import reduce

def square(x):
    return x**2

def parallel_square_sum(numbers):
    # Create a multiprocessing Pool
    with multiprocessing.Pool() as pool:
        # Use pool.map() for parallel processing
        squared_results = pool.map(square, numbers)
    #sum the squared results
    total_sum = reduce(lambda x, y: x + y, squared_results)
    return total_sum
````

## Exercise 7
Implement a MapReduce-style function to process data distributed across multiple nodes or workers.

````python
from functools import reduce

def map_reduce(data, mapper, reducer):
  intermediate_results = list(map(mapper, data))
  final_result = reduce(reducer, intermediate_results)
  return final_result

data = [1, 2, 3, 4, 5]
def mapper(x):
  return x ** 2
def reducer(x, y):
  return x + y

result = map_reduce(data, mapper, reducer)
print(result)
````
